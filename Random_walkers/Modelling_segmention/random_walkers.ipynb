```python
import numpy as np
import pylab as plt
```

```python
def x_positions(size,N):
    x_array = np.zeros((size,N))
    for ii in range(0,N):
        x = np.random.randint(-1,high = 2,size = size)
        xlist = []
        xcount = 0
        for i in range(0,size):
            xcount += x[i]
            xlist.append(xcount)
        p = np.array(xlist, dtype = int)
        x_array[:,ii] = p
    return x_array
            


def y_positions(size,N):
    y_array = np.zeros((size,N))
    for ii in range(0,N):
        y = np.random.randint(-1,high = 2,size = size)
        ylist = []
        ycount = 0
        for i in range(0,size):
            ycount += y[i]
            ylist.append(ycount)
        q = np.array(ylist, dtype = int)
        y_array[:,ii] = q
    return y_array

def def z_positions(size,N):
    z_array = np.zeros((size,N))
    for ii in range(0,N):
        z = np.random.randint(-1,high = 2,size = size)
        zlist = []
        zcount = 0
        for i in range(0,size):
            zcount += z[i]
            zlist.append(zcount)
        w = np.array(zlist, dtype = int)
        z_array[:,ii] = w
    return z_array
```

```python
positions_x = x_positions(30,30)
positions_y = y_positions(30,30)
positions_z = z_positions(30,30)
print(positions_x,positions_y)
```
